```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
import scipy.special
import seaborn as sns; sns.set_context("talk")


# creating a numpy array for x-axis
# using step size as 1
counts = np.arange(0, 10, 1)
mean=np.log(10)
# poisson distribution data for y-axis
L_discrete   = poisson.pmf(counts, mu=mean, loc=0)

x=np.linspace(0,10,100)
y=(np.exp(-mean) * mean**x)/scipy.special.factorial(x)

fig, ax = plt.subplots()

ax.set_title(f"Poisson distribution w/ expectation $\mu=${mean:.2f}")
ax.set_xlabel("n (counts)")
ax.set_ylabel("probability")

ax.bar(counts, L_discrete)
ax.plot(x,y,'r')

```
