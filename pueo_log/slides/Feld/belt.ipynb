```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_context("talk")

# Create a range of x values
x = np.linspace(0, 10, 500)

# Approximate upper and lower bounds for the region
y_upper = (x-1)**2
y_lower = (x-3)**2

# Create the plot
fig, ax = plt.subplots(1)

# ax.plot(x,y_lower)
# ax.plot(x,y_upper)

# Draw discrete horizontal lines closely spaced across the region
y_values = np.arange(0, 25, 0.5 )  
# Define the y-value where the red line will appear
red_line_y = 3.5

for y in y_values:
    # Find the corresponding x values where y is between y_lower and y_upper
    valid_indices = np.where((y >= y_lower) & (y <= y_upper))
    
    # Plot red line at the specific y-value, black lines otherwise
    if np.isclose(y, red_line_y):
        ax.plot(x[valid_indices], np.full_like(x[valid_indices], y), color='red', lw=1.5)
    else:
        ax.plot(x[valid_indices], np.full_like(x[valid_indices], y), color='black', lw=0.8)


# Add multiple vertical lines using plt.vlines
np.random.seed(1)
vline_x_values = np.random.uniform(3,5, size=4)  
plt.vlines(vline_x_values, ymin=0, ymax=25, colors='black', linestyles='--', lw=0.8)
plt.vlines(6, ymin=0, ymax=25, colors='black', linestyles='--', lw=0.8)


# # Set the x and y axis limits
ax.set_xlim([2, 8])
ax.set_ylim([0, 25])

# Add labels
ax.set_xlabel('x')
ax.set_ylabel(r'$\mu$')

# Customize the ticks and grid
ax.grid(False)
```
