```python
import numpy as np
from scipy.stats import rayleigh
import matplotlib.pyplot as plt
plt.rcParams.update({"font.size": 20})
```

```python
def fun(xlabel, ylabel, title):


    # generate sample data
    data = rayleigh.rvs(size=5000)


    x_min, x_max = rayleigh.ppf(0.01), rayleigh.ppf(0.999)

    x = np.linspace(x_min, x_max, 100)
    fig, ax = plt.subplots(figsize=(10,10))
    # histogram bars
    # ax.hist(data, bins=40, range = (x_min, x_max),density=True, alpha=0.4, edgecolor='black')

    ax.plot(x,rayleigh.pdf(x))
    ax.fill_between(x, rayleigh.pdf(x), alpha=0.4)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel, rotation=0, labelpad=43)
    ax.tick_params(axis='y',labelleft=False)
    # ax.tick_params(axis='x',labelbottom=False)
    # force same y range
    ax.set_ylim(0, rayleigh.pdf(x).max() * 1.1)
    fig.savefig(f"img/{title}.png")

# fun(xlabel="Energy [linear units]", ylabel="dN/dE", title="linear_dnde_bar")
fun(xlabel="Energy [linear units]", ylabel="E dN/dE", title="linear_EdNdE")
fun(xlabel="X [decade]", ylabel="dN/dX", title="log_dNdX")
fun(xlabel="X [decade]", ylabel="E dN/dE", title="log_EdNdE")
fun(xlabel="X [decade]", ylabel=r"$E^2$ dN/dE", title="log_E2dNdE")
```

```python

x = np.linspace(rayleigh.ppf(0.01), rayleigh.ppf(0.999), 100)
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,rayleigh.pdf(x))
ax.fill_between(x, rayleigh.pdf(x), alpha=0.4)
ax.set_xlabel("Energy [arbitray units]")
ax.set_ylabel("dN/dE", rotation=0, labelpad=43)
ax.tick_params(axis='y',labelleft=False)
# ax.tick_params(axis='x',labelbottom=False)
fig.savefig("img/linear_dnde.png")
```
